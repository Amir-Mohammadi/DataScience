# Summations (Σ)



```python
# this is the normal code to run summation

n = int(input())
summation = sum(10*i for i in range(1,n))
print(summation)
```

this is the sympy way 

```python
from sympy import * 
i,n = symbols('i n')

# iterate each element i from 1 to n

summation = Sum(i*i,(i,1,n))
upToFive = summation.subs(n,5)
print(upToFive.doit())
```

# Limits


```python
from sympy import *

x = symbols('x')
f = (1 + (1/x))**x
result = limit(f,x,oo)
# plot(f)
print("limit : ", result, "=", result.evalf())
```

# Derivatives And Integrate


```python

# normal method of derivative of a function

def derivativeFunction(f,x,stepSize):
    return (f(x + stepSize) - f(x))/ ((x + stepSize) - x)

def currentFunction(x):
    return x**2

```

```python
# sympy method

from sympy import *

x = symbols('x')
f = x**2

dx_f = diff(f)
diffs = dx_f.subs(x,2)
print(dx_f, diffs)
plot(f,dx_f)
```

```python
from sympy import *
from sympy.plotting import plot3d
x,y = symbols('x y')

f = 2*x**3 + 4*x**5 + 3*y**3


dx_f = diff(f,x)
dy_f = diff(f,y)

print('dx_f = ', dx_f,"|",'dy_f = ',dy_f)

plot3d(f)
```

```python
from sympy import *

x = symbols('x')

f = x**2



result = integrate(f,x)


print(result)
```

# Probality


```python
from scipy.stats import binom
n = 10
p = 0.9

for k in range(n+1):
    probability = binom.pmf(k,n,p)
    print("{0} - {1}".format(k,probability))
    
plot(probability)
```

```python
from scipy.stats import beta

a = 8
b = 2

p = 1.0 - beta.cdf(.90, a,b)

print(p)

```

```python
sample = [1,3,2,5,7,0,2,3]

mean = sum(sample)/ len(sample)

print(mean)



sample = [90, 80, 63, 87]
weights = [.20, .20, .20, .40]

weighted_mean = sum(s * w for s,w in zip(sample , weights)) / sum(weights)

print(weighted_mean)
```

```python


# Median 

sample=[0,1,5,7,9,10,14]


def median(values):
    ordered = sorted(values)
    print(ordered)
    n = len(ordered)
    mid = int(n/2) - 1 if n % 2 == 0 else int(n/2)
    
    if(n % 2 == 0):
        return(ordered[mid] + ordered[mid+1]) / 2
    else:
        return ordered[mid]

print(median(sample))
        
```
